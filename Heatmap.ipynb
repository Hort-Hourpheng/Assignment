```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import datetime as dt
#Load data
df=pd.read_excel("./Real estate valuation data set.xlsx")
df
```

```python
cr = df.corr(numeric_only=True)
sns.heatmap(cr, annot=True, cmap="coolwarm")
```

```python
matrix=np.triu(cr)
sns.heatmap(
    cr,
    annot=True,
    cmap="RdBu",
    vmin=-1,
    center=0,
    vmax=1,
    cbar=1,
    linewidths=2,
    square=True,
    mask=matrix,
    xticklabels=['X1','X2','X3','X4','X5','X6','X7','X8'],
    yticklabels="auto"
)
plt.title("heatmap of X1 and Y")
```

```python
col=df.columns.to_list()
```

```python
fig, hm=plt.subplots()
hm.set_xticklabels(col)
hm.set_yticklabels(col)
plt.pcolormesh(cr, cmap='RdBu',vmin=-1,vmax=1)
plt.title("Heatmap of X1 and Y")
plt.setp(hm.get_xticklabels(),rotation=45,ha="right",
         rotation_mode="anchor")
plt.setp(hm.get_yticklabels(),rotation=0,ha="right")
plt.colorbar()
plt.show()
```
